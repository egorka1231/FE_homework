Реализовать самостоятельно логистическую регрессию  
Обучить ее методом градиентного спуска  
Методом nesterov momentum  
Методом rmsprop  

В качестве dataset’а взять Iris, оставив 2 класса:  
Iris Versicolor  
Iris Virginica  

```python
import pandas as pd
from sklearn import datasets
import numpy as np
from matplotlib import pyplot as plt
```

```python
iris = datasets.load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data.head()
```

```python
data['target'] = iris.target
data
```

```python
data = data.loc[data['target'].isin([1, 2])]
data['target'].loc[data['target']==1]=0
data['target'].loc[data['target']==2]=1
data
```

```python
x = data.drop(columns=['target'])
y = data['target']
x
```

```python
y
```

```python
#Реализовать самостоятельно логистическую регрессию
theta = np.random.normal(size=(5,))
theta
```

```python
def predict(theta, x):
    predict = theta[0] + theta[1] * x.values[:, 0] + theta[2] * x.values[:, 1] + theta[3] * x.values[:, 2] + theta[4] * x.values[:, 3]
    sigm = 1. / (1 + np.exp(-predict))
    return sigm
```

```python
#Обучить ее методом градиентного спуска
lr = 0.1
grad_loss_hist = []
for _ in range(50):
    sigm = predict(theta, x)
    theta -= lr * np.sum(np.dot((sigm - y), x))/len(sigm)
    loss = - np.mean(np.log(sigm) * y + np.log(1 - sigm) * (1 - y))
    print(theta)
    grad_loss_hist.append(loss)
```

```python
#Методом nesterov momentum

lr = 0.05

vel_pred = np.zeros(5)
vel = np.zeros(5)
nest_loss_hist = []
gamma = 0.975
for _ in range(50):
    sigm = predict(theta, x)
    loss = - np.mean(np.log(sigm) * y + np.log(1 - sigm) * (1 - y))
    nest_loss_hist.append(loss)
    sigm = predict(theta - gamma * vel_pred, x)
    vel = np.dot(gamma, vel_pred) + lr * np.sum(np.dot((sigm - y),x))/len(sigm)

    theta -= vel
    print(theta)

    vel_pred = vel

```

```python
#Методом rmsprop 
lr = 0.1
e_sq_grad = np.zeros(5)
grad = np.zeros(5)
rmsp_loss_hist = []
gamma = 0.975
eps = 0.00000001
for _ in range(50):
    sigm = predict(theta, x)
    loss = - np.mean(np.log(sigm) * y + np.log(1 - sigm) * (1 - y))
    rmsp_loss_hist.append(loss)
    
    grad = np.sum(np.dot((sigm - y), x))/len(sigm)
    
    e_sq_grad = gamma * e_sq_grad + (1 - gamma)  * grad ** 2
    
    theta -= lr * grad / np.sqrt(e_sq_grad + eps)
    print(theta)
```

```python
plt.plot(range(50), grad_loss_hist, label='Градинтный спуск' )
plt.plot(range(50), nest_loss_hist, label='nesterov momentum' )
plt.plot(range(50), rmsp_loss_hist, label='rmsprop' )
plt.title('Скорость убывания ошибки')
plt.legend()
plt.show()

```
